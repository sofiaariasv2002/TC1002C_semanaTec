```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame
data = pd.read_csv('peso_estatura_genero.csv')

# Verificar la cantidad de datos y las variables
num_rows, num_cols = data.shape
column_names = data.columns
data_types = data.dtypes

print(f"El archivo contiene {num_rows} filas y {num_cols} columnas.")
print("Las variables en el archivo son:")
print(column_names)
print("Los tipos de datos de cada variable son:")
print(data_types)

# Análisis de las variables
genero = data['Genero']
estatura = data['Estatura']
peso = data['Peso']

# Resumen estadístico
estadisticas_genero = genero.describe()
estadisticas_estatura = estatura.describe()
estadisticas_peso = peso.describe()

# Conclusiones a partir de las estadísticas
media_estatura = estadisticas_estatura['mean']
mediana_estatura = estadisticas_estatura['50%']
desviacion_estandar_estatura = estadisticas_estatura['std']

media_peso = estadisticas_peso['mean']
mediana_peso = estadisticas_peso['50%']
desviacion_estandar_peso = estadisticas_peso['std']

print("\nResumen estadístico de la variable 'Estatura':")
print(estadisticas_estatura)
print("\nResumen estadístico de la variable 'Peso':")
print(estadisticas_peso)

print("\nConclusiones:")
print(f"La variable 'Estatura' tiene una media de {media_estatura:.2f}, una mediana de {mediana_estatura:.2f}, y una desviación estándar de {desviacion_estandar_estatura:.2f}.")
print(f"La variable 'Peso' tiene una media de {media_peso:.2f}, una mediana de {mediana_peso:.2f}, y una desviación estándar de {desviacion_estandar_peso:.2f}.")

# Análisis del género
conteo_genero = genero.value_counts()

print("\nConteo de la variable 'Genero':")
print(conteo_genero)

```
